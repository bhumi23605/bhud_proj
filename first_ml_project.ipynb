```python
#load data
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df

```

```python
#data preparation
y=df['logS']
y
x=df.drop('logS',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)   #train=70%of data, test=30%of data
x_train
```

```python
x_test
```

```python
#MODEL BUILDING - linear regression-Training the model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)


```

```python
#applying model to make predictions
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
#evaluate model performance-
from sklearn.metrics import mean_squared_error,r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)
lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
df1=pd.DataFrame(['linear regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
df1.columns=['model','train mse','train r2','test mse','test r2']
df1
```

```python
#MODEL BUILDING - Random forest- training the model
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

1) if logS i.e y varibale is categorical value then build a classification model
if logS is quantitative value then build a regression model,
2) here logs is quant therefore regressionn model.
3) The max_depth of a tree in Random Forest is defined as the longest path between the root node and the leaf node:

```python
#applying model to make predictions
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

```python
#evaluate model performance
from sklearn.metrics import mean_squared_error,r2_score
rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)
rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
df2=pd.DataFrame(['random forest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
df2.columns=['model','train mse','train r2','test mse','test r2']
df2
```

```python
# MODEL COMPARISON
df_models=pd.concat([df1,df2])
df_models.reset_index(drop=True)
```
